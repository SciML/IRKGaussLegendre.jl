# Lorenz Problem

```julia
using Pkg
using IRK, LinearAlgebra, OrdinaryDiffEq, Plots
```

```julia
function lorenz!(du,u,p,t)
du[1] = 10.0*(u[2]-u[1])
du[2] = u[1]*(28.0-u[3]) - u[2]
du[3] = u[1]*u[2] - (8/3)*u[3]
end
```

##  First:  Tsit5() algorithm 

```julia
u0 = [1.0;0.0;0.0]
tspan = (0.0,100.0)
prob = ODEProblem(lorenz!,u0,tspan)
sol = solve(prob,Tsit5());
```

```julia
plot(sol,vars=(1,2,3))
```

## Second: IRK8 algorithm

```julia
n=length(sol.t);
```

```julia
#dt=(tspan[2]-tspan[1])/n
#saveat=dt
(sol1,iters1,steps1) = solve(prob,IRK8(),dt=(tspan[2]-tspan[1])/n,
                             adaptive=false, myoutputs=true);
```

```julia
plot(sol1,vars=(1,2,3))
```
