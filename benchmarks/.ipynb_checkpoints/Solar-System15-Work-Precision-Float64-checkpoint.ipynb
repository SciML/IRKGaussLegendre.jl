## Initial value problem: Solar-System15 problem

The vector $Gm = (G\, m_1, \ldots, G\, m_N)$

```julia
include("../examples/InitialNBody15.jl")
u0, Gm =InitialNBody15()
N = length(Gm)
(size(u0), typeof(u0),typeof(Gm))
```

Just to check that the output of the function NbodyODE has the expected type:

```julia
include("../examples/Nbody.jl")
du0=similar(u0)
NbodyODE!(du0,u0,Gm,0.)
@show size(du0);
```

```julia
t0 = 0.
T = 100. # Final time

q0=u0[2,:,:]
v0=u0[1,:,:]

tspan=(t0,T)
prob=ODEProblem(NbodyODE!,u0,tspan,Gm);
probdyn = DynamicalODEProblem(NbodyODEq!,NbodyODEv!,q0,v0,tspan,Gm);
```

```julia
setprecision(BigFloat, 106)
u0128, Gm =InitialNBody15(BigFloat)
tspan128=(BigFloat(0.0),BigFloat(T))
prob128=ODEProblem(NbodyODE!,u0128,tspan128,Gm);
```

## Integration with Vern9

Recommend methods: Non-Stiff Problems

- For high accuracy non-stiff solving ( BigFloat and tolerances like <1e-12 ), JuliaDiffeq recommend
the Vern9 method


```julia
#sol0 = solve(prob128,Vern9(),saveat=10,abstol=1e-20,reltol=1e-20);
#@save "./Data/solarsystem15small_test0_solF64.jld2" sol0
@load "./Data/solarsystem15small_test0_solF64.jld2" sol0
test_sol = TestSolution(sol0);
```

```julia
sol0.destats
```

## Integration with IRKGL16 (Adaptive-step)


```julia
(sol1,iter1,steps1)=solve(prob,IRKGL16(),
                               reltol=1e-12, abstol=1e-12,myoutputs=true,initial_interp=true);
```

```julia
sol1.destats
```

```julia
(sol2,iter2,steps2)=solve(probdyn,IRKGL163(), reltol=1e-12, abstol=1e-12,myoutputs=true,initial_interp=true);
sol1.destats
```

```julia
plot(sol1.t[2:end],iter1[2:end], title="Iteration numbers", legend=false)
```

```julia
plot(sol1.t,steps1, title="step sizes in the integration", legend=false)
plot!(sol2.t,steps2)
```

## Integration with IRKGL16 (Fixed-step)

```julia
dt = 10 # time-step length
(sol3,iter3,steps3)=solve(prob,IRKGL16(),adaptive=false,dt=dt,myoutputs=true,initial_interp=true)
sol3.destats
```

```julia
(sol4,iter4,steps4)=solve(probdyn,IRKGL163(),adaptive=false,dt=dt,myoutputs=true,initial_interp=true)
sol4.destats
```

## Error in energy

```julia
setprecision(BigFloat, 256)
u0128, Gm128 =InitialNBody15(BigFloat)
E0=NbodyEnergy(u0128,Gm128)

ΔE0 = map(x->NbodyEnergy(BigFloat.(x),Gm128), sol0.u)./E0.-1
ΔE1 = map(x->NbodyEnergy(BigFloat.(x),Gm128), sol1.u)./E0.-1
ΔE2 = map(x->NbodyEnergy2(BigFloat.(x),Gm128), sol2.u)./E0.-1
ΔE3 = map(x->NbodyEnergy(BigFloat.(x),Gm128), sol3.u)./E0.-1
ΔE4 = map(x->NbodyEnergy2(BigFloat.(x),Gm128), sol4.u)./E0.-1

(Float32(maximum(abs.(ΔE0))),Float32(maximum(abs.(ΔE1))),Float32(maximum(abs.(ΔE2))),Float32(maximum(abs.(ΔE3))),
Float32(maximum(abs.(ΔE4))))
```

```julia
plot(sol0.t,log10.(abs.(ΔE0)), label="Test solution")
```

```julia
plot(sol1.t,log10.(abs.(ΔE1)), label="IRKGL16 (Adaptive-ODEProblem)")
plot!(sol2.t,log10.(abs.(ΔE2)), label="IRKGL16 (Adaptive-Dynamic)")
```

```julia
plot(sol3.t,log10.(abs.(ΔE3)), label="IRKGL16 (Fixed-ODEProblem)")
plot!(sol4.t,log10.(abs.(ΔE4)), label="IRKGL16 (Fixed-Dynamic)")
```

## Work-Precision

```julia
abstols = 1.0 ./ 10.0 .^ (8:14)
reltols = 1.0 ./ 10.0 .^ (8:14);
dts=20.0./2.0.^(0:length(reltols)-1)
```

```julia
setups = [
#    Dict(:alg=>DP8())
          Dict(:alg=>Vern9())
#          Dict(:alg=>Vern8())   
          Dict(:alg=>IRKGL16())
          Dict(:alg=>IRKGL16(),:adaptive=>false,:dts=>dts)       
    ];
wp = WorkPrecisionSet(prob,abstols,reltols,setups;appxsol=test_sol,save_everystep=false,numruns=1)
plot(wp)
```
